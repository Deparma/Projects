```python
# Взвешенное скользящее среднее / Weighted moving average
# Функция взвешенного скользязщего среднего, веса нормируются в коде
def wma(x, n, weights):
    weights = np.array(weights)
    wmas = x.rolling(n).apply(lambda x: np.dot(x, weights) /
                                       weights.sum(), raw=True).to_list()
    return wmas
```

```python
plt.plot(df.index, df['Temp'],
         linewidth=2, label='actual', alpha=.7)
plt.plot(df.index, wma(df['Temp'], n=3, weights=[1, 5, 10]), 
         linewidth=2, label=f'n=3, weights=[1, 5, 10]', alpha=.7)
plt.plot(df.index, wma(df['Temp'], n=6, weights=[10, 8, 5, 3, 2, 1]),
         'r', linewidth=2, label='n=6, weights=[10, 8, 5, 3, 2, 1]', alpha=.5)
plt.legend()
```

```python
# Экспоненциальное сглаживание / Exponential smoothing
# Простое экспоненциальное сглаживание
# Функция простого экспоненциального сглаживания
def exponential_smoothing(series, alpha):
    result = [series[0]]
    for n in range(1, len(series)):
        result.append(alpha * series[n] + (1 - alpha) * result[n-1])
    return result
```

```python
plt.figure(figsize=(14,5))
plt.plot(df.index, df['Temp'], linewidth=2, label='actual', alpha=.7)

# Применяем сглаживание с разными параметрами и записыем в переменные
exp_sm_5 = exponential_smoothing(df['Temp'], 0.5)
exp_sm_2 = exponential_smoothing(df['Temp'], 0.2)
exp_sm_05 = exponential_smoothing(df['Temp'], 0.05)

# Строим графики
plt.plot(df.index, exp_sm_5, 'r', linewidth=2, label='alpha 0.5', alpha=.5)
plt.plot(df.index, exp_sm_2, 'g', linewidth=2, label='alpha 0.2', alpha=.5)
plt.plot(df.index, exp_sm_05, 'orange', linewidth=2, label='alpha 0.05')
plt.legend();
```

```python
# Функция двойного экспоненциального сглаживания
def double_exponential_smoothing(series, alpha, beta):
    result = [series[0]]
    for n in range(1, len(series)+1):
        if n == 1:
            level, trend = series[0], series[1] - series[0]
        if n >= len(series): 
            value = result[-1]
        else:
            value = series[n]
        last_level, level = level, alpha*value + (1-alpha)*(level+trend)
        trend = beta*(level-last_level) + (1-beta)*trend
        result.append(level+trend)
    return result
```

```python
dbl_exp_sm_08_08 = double_exponential_smoothing(df['Temp'], .08, .08)[1:]
dbl_exp_sm_35_35 = double_exponential_smoothing(df['Temp'], .35, .35)[1:]
dbl_exp_sm_45_45 = double_exponential_smoothing(df['Temp'], .45, .45)[1:]

plt.plot(df.index, dbl_exp_sm_08_08, 'r', linewidth=2, label='alpha 0.08 beta 0.08', alpha=.4)
plt.plot(df.index, dbl_exp_sm_35_35, 'g', linewidth=2, label='alpha 0.2 beta 0.2', alpha=.4)
plt.plot(df.index, dbl_exp_sm_45_45, 'orange', linewidth=2, label='alpha 0.8 beta 0.8', alpha=.4)
plt.legend();
```

```python
from tsfresh import extract_features
```

```python
df_1 = pd.read_csv('daily-min-temperatures.csv')
```

```python
# генерация признаков из временных рядов
extracted_features = extract_features(df_1, column_id='Date', n_jobs=4)
```

```python
# избавление от неинформативных столбоцв и редактирование оставшихся нанов
null_cols = []
for col in extracted_features.columns:
    nans_percent = sum(extracted_features[col].isna()) / len(extracted_features) * 100
    if extracted_features[col].sum() == 0 or nans_percent > 10:
        null_cols.append(col)

df_ext_2 = extracted_features.drop(null_cols, axis=1)   
        
for col in df_ext_2.columns:
    if df_ext_2[col].isna().sum() != 0:
        df_ext_2[col] = df_ext_2[col].fillna(df_ext_2[col].mean())
```

```python
df_ext_2
```

```python
from sklearn.model_selection import TimeSeriesSplit
```

```python
# разбиваем дф на обучение и тест
X = df_ext_2.values
y = df.Temp
```

```python
y
```

```python
tss = TimeSeriesSplit(10)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# находим mse в линейной регрессии
lr = LinearRegression()
```

```python
train_mse = []
test_mse = []
for train_index, test_index in tss.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    
    lr.fit(X_train, y_train)
    
    mse_train = mean_squared_error(y_train, lr.predict(X_train))
    mse_test = mean_squared_error(y_test, lr.predict(X_test))
    print(f'Train mse: {mse_train}')
    print(f'Test mse: {mse_test}')
    print('-------------------------')
    
    train_mse.append(mse_train)
    test_mse.append(mse_test)
```

```python
plt.plot(train_mse)
plt.plot(test_mse)
```
